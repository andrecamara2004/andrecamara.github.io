# ðŸ“Š About the Dataset


**Dataset Overview:**  
This dataset mimics actual healthcare records, generated using Pythonâ€™s Faker library to provide realistic yet completely synthetic patient data. Hereâ€™s a quick breakdown of the dataset:
  
  - **Name:** Patient name.   
  
  - **Age:** Patient's age at admission.
  - **Gender:** Patient's gender.

  - **Discharge Date:** Date the patient was discharged.  
  - **Medication:** Medication prescribed to the patient.  
  - **Test Results:** Outcome of a medical test (e.g., "Normal," "Abnormal").

```python
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [16, 4]
```

```python
plt.hist(df['Age'], bins=40, edgecolor='black')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting billing amount distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Billing Amount'], bins=30, edgecolor='black')
plt.title('Billing Amount Distribution')
plt.xlabel('Billing Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting Admission Type counts
plt.figure(figsize=(10, 6))
df['Admission Type'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Admission Type Counts')
plt.xlabel('Admission Type')
plt.ylabel('Frequency')
plt.ylim(17000, 19000)  # Setting the y-axis range to improve visibility of differences

plt.show()
```

```python
# Calculating the average billing amount for each admission type
billing_by_admission_type = df.groupby('Admission Type')['Billing Amount'].mean()

# Plotting the average billing amount for each admission type
plt.figure(figsize=(10, 6))
billing_by_admission_type.plot(kind='bar', edgecolor='black')
plt.title('Average Billing Amount by Admission Type')
plt.xlabel('Admission Type')
plt.ylabel('Average Billing Amount')
plt.ylim(25000, 26000)  # Setting the y-axis range to improve visibility of differences
plt.show()
```

```python
billing_by_admission_type
```

```python
# Calculating the average billing amount for each medical condition
billing_by_medical_condition = df.groupby('Medical Condition')['Billing Amount'].mean().sort_values(ascending=False)

# Plotting the average billing amount for each medical condition (showing top 10 for readability)
plt.figure(figsize=(12, 8))
billing_by_medical_condition.head(10).plot(kind='bar', edgecolor='black')
plt.title('Medical Conditions by Average Billing Amount')
plt.xlabel('Medical Condition')
plt.ylabel('Average Billing Amount')
plt.ylim(25000, 26000)  # Setting the y-axis range to improve visibility of differences
plt.show()
```

```python
# Plotting a scatter plot to visualize the trend between Age and Billing Amount
plt.figure(figsize=(12, 8))
plt.scatter(df['Age'], df['Billing Amount'], alpha=0.3)
plt.title('Billing Amount vs Age')
plt.xlabel('Age')
plt.ylabel('Billing Amount')
plt.show()
```

```python
# Calculating the correlation between Age and Billing Amount to quantify the trend
age_billing_corr = df[['Age', 'Billing Amount']].corr().iloc[0, 1]
print("Correlation between Age and Billing Amount:", age_billing_corr)
```

```python
# Converting admission and discharge dates to datetime format if not already done
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])

# Calculating length of stay as the difference in days between admission and discharge
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
```

```python
# Calculating the average length of stay for each medical condition
avg_length_of_stay_by_condition = df.groupby('Medical Condition')['Length of Stay'].mean().sort_values(ascending=False)
avg_length_of_stay_by_condition
```

```python

# Displaying the top 10 medical conditions by average length of stay for better readability
plt.figure(figsize=(12, 8))
avg_length_of_stay_by_condition.head(10).plot(kind='bar', edgecolor='black')
plt.title('Medical Conditions by Average Length of Stay')
plt.xlabel('Medical Condition')
plt.ylabel('Average Length of Stay (Days)')
plt.ylim(14, 16)  # Setting the y-axis range to improve visibility of differences
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Filter for cancer cases
cancer_df = df[df['Medical Condition'] == 'Cancer']

# Convert the admission date to datetime
cancer_df['Date of Admission'] = pd.to_datetime(cancer_df['Date of Admission'])

# Count the number of cancer cases per day
cancer_cases_per_day = cancer_df['Date of Admission'].value_counts().sort_index()

# Prepare the data for regression
X = cancer_cases_per_day.index.map(lambda date: date.toordinal()).values.reshape(-1, 1)
y = cancer_cases_per_day.values

# Initialize and fit the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions on the training data for plotting
y_pred = model.predict(X)

# Plotting the actual cancer cases and the regression line
plt.figure(figsize=(12, 6))
plt.plot(cancer_cases_per_day.index, y, label='Actual Cancer Cases')
plt.plot(cancer_cases_per_day.index, y_pred, color='red', linestyle='--', label='Trend Line (Linear Regression)')
plt.title('Cancer Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cancer Cases')
plt.legend()
plt.show()




```
